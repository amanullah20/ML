```python
x = df.drop('rent',axis=1)
```

```python
x.head()
```

```python
y=df.drop('area',axis=1)
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.predict([[3000]])
```

```python
reg.predict(x)
```

```python
df['predicted_result']=reg.predict(x)
```

```python
df.head()
```

```python
plt.plot(df.area,df.predicted_result)

plt.scatter(df.area,df.rent)
plt.xlabel('Area in Square')
plt.ylabel('Rent in Bangladeshi Taka')
plt.title('Predicted_Area Rent plot')
```

```python
reg.score(x,y)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(df.rent,df.predicted_result)
```
